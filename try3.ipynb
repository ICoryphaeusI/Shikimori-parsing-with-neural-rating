```python
import pandas as pd

# Загрузка данных из файла CSV в DataFrame
combined_df = pd.read_csv('combined_data.csv')

# Теперь у вас есть DataFrame combined_df с данными из файла CSV
```

```python
combined_df
```

```python
# Указываем ненужные столбцы для удаления
columns_to_drop = ['Description']

# Удаляем столбцы
df = combined_df.drop(columns=columns_to_drop)
```

```python
df
```

```python
df.rename(columns={'21.1': '21'}, inplace=True)
```

```python
# Перемешиваем строки в DataFrame
df_shuffled = df.sample(frac=1, random_state=42)
```

```python
# Разделение данных на числовые значения, целевую переменную и закодированные тексты
numerical_df = df_shuffled.loc[:, 'Episodes_scaled':'Фумитоси Миядзима']
y = df_shuffled['Rate']
encoded_texts_df = df_shuffled.loc[:, '0':'599']
```

```python
encoded_texts_df
```

```python
from sklearn.model_selection import train_test_split

# Разделение данных на обучающую и тестовую выборки (90% обучающая, 10% тестовая)
numerical_train, numerical_test, y_train, y_test, encoded_texts_train, encoded_texts_test = train_test_split(
    numerical_df, y, encoded_texts_df, test_size=0.1, random_state=42)

# Разделение обучающей выборки на обучающую и валидационную выборки (90% обучающая, 10% валидационная)
numerical_train, numerical_val, y_train, y_val, encoded_texts_train, encoded_texts_val = train_test_split(
    numerical_train, y_train, encoded_texts_train, test_size=0.1, random_state=42)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate, BatchNormalization, ReLU, Dropout

# Вход для числовых значений
num_input = Input(shape=(numerical_df.shape[1],), name='numerical_input')
num_input1 = Dense(64, name='num_dense')(num_input)
num_input2 = BatchNormalization(name='num_batch_normalization')(num_input1)
num_input3 = ReLU(name='num_re_lu')(num_input2)
num_input4 = Dropout(0.5, name='num_dropout')(num_input3)
num_output = Dense(1, name='num_output')(num_input4)

# Вход для закодированных текстов
text_input = Input(shape=(encoded_texts_df.shape[1],), name='text_input')
text_input1 = Dense(32, name='text_dense')(text_input)
text_input2 = BatchNormalization(name='text_batch_normalization')(text_input1)
text_input3 = ReLU(name='text_re_lu')(text_input2)
text_input4 = Dropout(0.5, name='text_dropout')(text_input3)
text_input5 = Dense(1024, name='text_dense_1')(text_input4)
text_input6 = BatchNormalization(name='text_batch_normalization_1')(text_input5)
text_input7 = ReLU(name='text_re_lu_1')(text_input6)
text_input8 = Dropout(0.5, name='text_dropout_1')(text_input7)
text_input9 = Dense(32, name='text_dense_2')(text_input8)
text_input10 = BatchNormalization(name='text_batch_normalization_2')(text_input9)
text_input11 = ReLU(name='text_re_lu_2')(text_input10)
text_input12 = Dropout(0.5, name='text_dropout_2')(text_input11)
text_output = Dense(1, name='text_output')(text_input12)

concatenated_output = Concatenate()([num_output, text_output])

# Define regression_head_1 layer
regression_head = Dense(1, name='regression_head')(concatenated_output)
# Define the model
model = Model(inputs=[num_input, text_input], outputs=regression_head)

# Компиляция модели
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Вывод структуры модели
model.summary()
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate, BatchNormalization, ReLU, Dropout

# Вход для числовых значений
num_input = Input(shape=(numerical_df.shape[1],), name='numerical_input')
num_input1 = Dense(64, name='num_dense')(num_input)
num_input2 = BatchNormalization(name='num_batch_normalization')(num_input1)
num_input22 = Dense(512, name='num_dense512')(num_input2)
num_input3 = ReLU(name='num_re_lu')(num_input22)
num_input4 = Dropout(0.5, name='num_dropout')(num_input3)
num_input5 = Dense(1024, name='num_dense1024')(num_input4)
num_output = Dense(1, name='num_output')(num_input5)

# Вход для закодированных текстов
text_input = Input(shape=(encoded_texts_df.shape[1],), name='text_input')
text_input1 = Dense(32, name='text_dense')(text_input)
text_input2 = BatchNormalization(name='text_batch_normalization')(text_input1)
text_input3 = ReLU(name='text_re_lu')(text_input2)
text_input4 = Dropout(0.5, name='text_dropout')(text_input3)
text_input5 = Dense(1024, name='text_dense_1')(text_input4)
text_input6 = BatchNormalization(name='text_batch_normalization_1')(text_input5)
text_input7 = ReLU(name='text_re_lu_1')(text_input6)
text_input8 = Dropout(0.5, name='text_dropout_1')(text_input7)
text_input9 = Dense(1024, name='text_dense_2')(text_input8)
text_input10 = BatchNormalization(name='text_batch_normalization_2')(text_input9)
text_input11 = ReLU(name='text_re_lu_2')(text_input10)
text_input12 = Dropout(0.5, name='text_dropout_2')(text_input11)
text_output = Dense(1, name='text_output')(text_input12)

concatenated_output = Concatenate()([num_output, text_output])

# Define regression_head_1 layer
regression_head = Dense(1, name='regression_head')(concatenated_output)
# Define the model
model = Model(inputs=[num_input, text_input], outputs=regression_head)

# Компиляция модели
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Вывод структуры модели
model.summary()
```

```python
import matplotlib.pyplot as plt

# Обучаем модель
history = model.fit(x=[numerical_train, encoded_texts_train], y=y_train,
                    validation_data=([numerical_val, encoded_texts_val], y_val),
                    epochs=20, batch_size=32)


# Вывод графиков
# Получение значений loss и метрик из истории обучения
train_loss = history.history['loss']
val_loss = history.history['val_loss']
train_metric = history.history['mae']  # Предполагаем, что используется MAE как метрика
val_metric = history.history['val_mae']

# Создание графика для loss
plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Создание графика для метрики
plt.plot(train_metric, label='Training MAE')
plt.plot(val_metric, label='Validation MAE')
plt.title('Training and Validation MAE')
plt.xlabel('Epoch')
plt.ylabel('MAE')
plt.legend()
plt.show()
```

```python
# Оцениваем модель на тестовых данных
test_loss, test_mae = model.evaluate([numerical_test, encoded_texts_test], y_test)
print(test_loss)
print(test_mae)
```
