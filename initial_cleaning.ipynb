```python
import pandas as pd

df = pd.read_excel('data/data.xlsx')
df.head(10)
```

```python
print(df.info())
print('-'*100)
print(f'Размер датафрейма: {df.shape}')
```

```python
df = df[df['Rate'] != 'Не смог'].dropna(subset=['Rate'])
```

```python
print(df.info())
print('-'*100)
print(f'Размер датафрейма: {df.shape}')
```

```python
import matplotlib.pyplot as plt

# Преобразование столбца 'Rate' в числовой формат
df['Rate'] = pd.to_numeric(df['Rate'])

# Построение гистограммы распределения значений 'Rate' с подписями бинов
plt.figure(figsize=(25, 8))
n, bins, patches = plt.hist(df['Rate'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Распределение значений Rate')
plt.xlabel('Rate')
plt.ylabel('Частота')
plt.grid(True)

# Добавление подписей к бинам
for i in range(len(bins) - 1):
    plt.text(bins[i] + 0.05, n[i] + 0.5, str(round(bins[i], 2)), fontsize=8)

plt.xticks([])
plt.show()
```

```python
# Преобразование значений в столбце 'Episodes' в строки и фильтрация по наличию символа '/'
df = df[~df['Episodes'].astype(str).str.contains('/')]

# Вывод первых 10 строк
print(df.shape)
```

```python
df['Episodes'] = df['Episodes'].replace('Не указано', 1)
```

```python
mask = df['Episode length'] == '1 ча'
df.loc[mask, 'Episode length'] = '1 час'
```

```python
# Сохранение DataFrame в Excel файл
df.to_excel('initial_cleaning.xlsx', index=False)
```
